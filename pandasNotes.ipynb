# **Series**

---



```python
a = pd.Series([1,2,3,4,5,6,7,8])
print(a)
print(type(a))
print(a[3])
```

```python
#to change the index
a = pd.Series([1,2,3,4,5,6,7,8],index=['a','b','c','d','e','f','g','h'])
print(a)
print()
b = pd.Series([1,2,3,4,5,6,7,8],index=[1,2,3,4,5,6,7,8],dtype='float',name="Python Pandas")
print(b)
```

```python
#by using dictionary

dic = {
    "name" : "Saurabh Pal",
    "subject" : ["Java","Python","C","C++"],
    "roll_no" : 123,
    "marks" : 100
}
a = pd.Series(dic)
a
# print(a)
```

```python
a = pd.Series(12,index=[1,2,3,4,5,6,7,8])
print(a)
```

```python
#In numpy if we add two differnt array it will show the broadcast error but in pandas it will return NaN

a = pd.Series([1,2,3,4,5,6,7,8])
b = pd.Series([1,2,3,4,5])
print(a+b)
```

# **Arithmetic Operations in Pandas**



```python
a = pd.DataFrame({"A":[1,2,3,4,5,6,7,8],"B":[18,27,16,35,14,23,12,31]})
a
a["c"] = a["A"] + a["B"]
a
```

```python
a["c"] = a["A"] - a["B"]
a
```

```python
a["c"] = a["A"] * a["B"]
a
```

```python
a["c"] = a["A"] / a["B"]
a
```

```python
a["c"] = a["A"] %  a["B"]
a
```

```python
x = pd.DataFrame({"y":[1,2,3,4,5,6,7,8],"z":[18,27,16,35,14,23,12,31]})
x["Python"] = x["y"] > x["z"]
x["Python1"] = x["y"] > 6
x
```

# **Delete and Insert Data in Pandas**

```python
#insert(index_number_position_toInsert, new_column_position_name, value)

var = pd.DataFrame({"A":[1,2,3,4,5,6,7,8],"B":[18,27,16,35,14,23,12,31]})
var.insert(1,"Java",var["A"] + 2)
var
```

```python
var = pd.DataFrame({"A":[1,2,3,4,5,6,7,8],"B":[18,27,16,35,14,23,12,31]})
var.insert(1,"Java",var["A"][:4])
var
```

```python
#Delete

var = pd.DataFrame({"A":[1,2,3,4,5,6,7,8],"B":[18,27,16,35,14,23,12,31],"C":[1,2,3,4,3,2,3,4]})
var.pop("A")
var
```

```python
del var["B"]
var
```

# **Create CSV(Comma Seprated Value) File using Pandas**

```python
dic = {
    "Name":["Saurabh","Ayush","Bhavesh","Chandresh","Dolly"],
    "Surname":["Pal","Pal","Sharma","Tiwari","Mishra"],
    "Age": [23,13,16,26,18]
}
var = pd.DataFrame(dic)
var
```

```python
# a.to_csv("csvFile.csv",index = "",header=[])
# if index = True than tha index file will we shown to us otherwise it will be not created.
# header is used to give the name of the column.
var.to_csv("Test.csv",index = False)
```

```python
# read_csv("path_or_fileName",nrows=2 , usecols=[],skiprows=[],index_col="",header=2/None,names=[])
# nrows = return the rows
# usecols = return the columns
# skiprows = skip the rows
# index_col = to make any column as index
# header = is used to give the header of any row to the table if it is none than no header in the file.
# names = to give the name of the header to the columns
a = pd.read_csv("Test.csv",nrows=3,usecols=["Name","Age"])
a

```
